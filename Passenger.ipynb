```python
import pandas as pd 

ulr = 'passenger.csv'
data = pd.read_csv(ulr)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot = True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Miles(X)']), y = data['Dollars (Y)'])
print('he so intercep la: ',model.intercept_)
print('he so coeff', model.coef_)
model.score(X=pd.DataFrame(data['Miles(X)']), y = data['Dollars (Y)'])
```

```python
y_pred = model.predict(X=pd.DataFrame(data['Miles(X)']))
data.plot(kind = 'scatter',x = 'Miles(X)', y = 'Dollars (Y)', figsize = (9,9), color = 'black', xlim = (0.7))
plt.plot(data['Miles(X)'], y_pred,)
print('he so intercep la: ',model.intercept_)
print('he so coeff', model.coef_)
model.score(X=pd.DataFrame(data['Miles(X)']), y = data['Dollars (Y)'])
```

```python
import numpy as np
model=linear_model.LinearRegression()
X=pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2,data['Miles(X)']**3,data['Miles(X)']**4,data['Miles(X)']**5,data['Miles(X)']**6,data['Miles(X)']**7,data['Miles(X)']**8,data['Miles(X)']**9,data['Miles(X)']**10]).T
model.fit(X,y=data['Dollars (Y)'])

range=np.arange(0,6000,1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='Miles(X)',y='Dollars (Y)',figsize=(9,9),color='black',xlim=(0.7))
plt.plot(range,y_pred,color='blue')
print('Intercept của mô hình:',model.intercept_)
print('coeffient của  mô hình:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X,data['Dollars (Y)']))
```

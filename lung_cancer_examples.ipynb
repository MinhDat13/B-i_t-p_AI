```python
import pandas as pd 
ulr = 'lung_cancer_examples.csv'
data = pd.read_csv(ulr)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
correlation = data.corr()
```

```python
import seaborn as sns
sns.heatmap(correlation,annot = True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Smokes']), y = data['Result'])
print('he so intercep la: ',model.intercept_)
print('he so coeff', model.coef_)
model.score(X=pd.DataFrame(data['Smokes']), y = data['Result'])
```

```python
y_pred = model.predict(X=pd.DataFrame(data['Smokes']))
data.plot(kind = 'scatter',x = 'Smokes', y = 'Result', figsize = (9,9), color = 'black', xlim = (0.7))
plt.plot(data['Smokes'], y_pred,)
```

```python
model = linear_model.LinearRegression()
import numpy as np
X=pd.DataFrame([data['Smokes'],data['Smokes']**2,data['Smokes']**3]).T
model.fit(X, y = data['Result'])
range=np.arange(0,35,0.1)
X_range=pd.DataFrame([range,range**2,range**3]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='Smokes',y='Result',figsize=(9,9),color='black',xlim=(0.7))
plt.plot(range,y_pred,color='blue')
print('intercept:', model.intercept_)
print('coeff:', model.coef_)
print('do chinh xac', model.score(X,y = data['Result']))
```
